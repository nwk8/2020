# PART 1 – COMPLETE THEORY REVIEW

## 1. OOP Concepts

### Encapsulation
- Hiding internal data using private variables
- Access through public methods or properties
- Prevents invalid state

### Abstraction
- Hiding implementation details
- Showing only essential behavior

### Inheritance
- Child class derives from parent using :
- Promotes code reuse

### Polymorphism
- Method overriding
- One interface, multiple implementations

---

## 2. Data Structures

### Stack (LIFO)
- Push()
- Pop()
- Peek()
- O(1) operations (amortized for arrays)

### Queue (FIFO)
- Enqueue()
- Dequeue()
- Circular array recommended

### Singly Linked List
- Node contains: data + next
- Head pointer required

### Doubly Linked List
- Node contains: data + next + prev
- Head and Tail pointers

### Arrays
- O(1) access
- O(n) insertion/deletion due to shifting
- Fixed size unless resized

---

## 3. Time Complexity

O(1) → Constant  
O(n) → Linear  
O(log n) → Logarithmic  
O(n²) → Quadratic
